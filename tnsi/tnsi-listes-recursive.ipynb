# Listes définies «&nbsp;récursivement&nbsp;»

### L'interface :
Définissons une liste de façon récursive comme étant :
- Soit une liste vide
- Soit une paire dont le premier élément est une donnée et le deuxième élément est une liste

Et choisissons comme opérations primitives :
- `construit(e, L)` pour construire une nouvelle liste contenant un premier élément `e` (sa tête) et une suite `L` (sa queue, qui est une liste).
- `premier(L)` pour accéder au premier élément de la liste `L`. 
- `suite(L)` pour accéder à la suite de la liste `L`.
- `estvide(L)` pour tester si une liste est vide.
