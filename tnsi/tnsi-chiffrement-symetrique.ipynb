### Attaque :

Avec cette méthode, on voit que les décalages peuvent aller de 1 à 25. Il n'y a donc que 25 possibilités de codage différent. Ce n'est pas beaucoup !  

On décide d'attaquer ce chiffrement par **force brute**, c'est à dire en testant tous les décalages possibles et en choisissant la réponse qui a du sens en français.  

```python
def force_brute_césar(message):
    for décalage in range(1,26):
        print(chiffrement_césar(message, décalage))
```

## 4. Le chiffrement avec XOR
### Principe :
Un ordinateur ne manipule que des nombres. Avec l'avènement de l'informatique, on va donc convertir les lettres en nombres binaires, par exemple en prenant leur code UTF8, puis on va chiffrer la suite des nombres binaires obtenus. 
On garde donc le principe très efficace d'une clé longue et changée régulièrement, mais :
- au lieu de travailler sur des lettres, on travaille sur des nombres binaires.
- au lieu d'ajouter des décalages de caractères, on va faire des opérations binaires sur ces nombres.

Or parmi les opérations binaires, le XOR (⊕) présente deux avantages remarquables : 
- c'est une opération très rapide car ces portes font partie des portes directement "cablées" dans les UAL des processeurs.
- si a ⊕ b = c, alors c ⊕ b = a. Ce qui veut dire qu'en rechiffrant un message chiffré avec la même clé, on retrouve le message initial !
