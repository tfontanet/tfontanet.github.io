```python
import graphviz

dot = graphviz.Graph()
graphe = {"A":("B","F"),
          "B":("A","C","D","G"),
          "C":("B","E"),
          "D":("B","I"),
          "E":("C","I"),
          "F":("A","G","H"),
          "G":("B","F","I"),
          "H":("F","I"),
          "I":("D","E","G","H")}

# Création des noeuds
for noeud in graphe:
    dot.node(str(noeud))
    
# Création des arêtes : ("a", "b") et ("b", "a") ne doivent pas s'y trouver en même temps
arêtes = []
for noeud in graphe:
    for elem in graphe[noeud]:
        if (str(elem), str(noeud)) not in arêtes:
            arêtes.append((str(noeud), str(elem))) 
dot.edges(arêtes)        

# Graphique
graphviz.Source(dot)
```
