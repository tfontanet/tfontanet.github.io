```python
class Noeud():
    def __init__(self, clé, gauche = None, droit = None):
        self.clé = clé
        self.gauche = gauche
        self.droit = droit
    
    def __repr__(self):
        return f"({self.clé} -> {self.gauche}, {self.droit})".replace("None"," ")
```
